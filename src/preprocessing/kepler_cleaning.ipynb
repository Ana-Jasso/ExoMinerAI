```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/Kepler_cumulative_2025.10.04_12.41.25.csv')
df.head()  # Muestra las primeras filas para verificar la carga
```

```python
df.shape  # Muestra la forma del DataFrame
```

```python
df.isnull().sum()
```

```python
# Elimina filas con mÃ¡s de 3 valores nulos
umbral = 3
df_filtrado = df[df.isnull().sum(axis=1) <= umbral]
df_filtrado.shape  # Muestra la nueva forma del DataFrame
```

```python
df_filtrado.isnull().sum()
```

```python
df_filtrado = df_filtrado.drop(columns=[
    'koi_teq_err1', 'koi_teq_err2'
]) #elimina columnas con muchos valores nulos
```

```python
df_filtrado.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reemplaza 'target' por el nombre real de tu columna objetivo, por ejemplo 'koi_disposition'
correlacion = df_filtrado.corr(numeric_only=True)

plt.figure(figsize=(15, 10))
sns.heatmap(correlacion, annot=False, cmap='coolwarm')
plt.title('Mapa de calor de correlaciÃ³n entre variables y el target')
plt.show()
```

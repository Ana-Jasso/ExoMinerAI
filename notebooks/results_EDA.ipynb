```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master_df = pd.read_csv('.././data/processed/cleaned_datasets.csv')
```

```python
# Numeric columns: histograms
master_df.hist(figsize=(16, 12), bins=30)
plt.tight_layout()
plt.show()

# Categorical columns: bar plots
# Histogramas para columnas numéricas usando seaborn
numeric_cols = master_df.select_dtypes(include='number').columns
for col in numeric_cols:
    sns.histplot(master_df[col], bins=30, kde=True, color='blue')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(master_df[numeric_cols].corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Mapa de calor de correlación entre variables numéricas')
plt.show()
```

```python
correlation_matrix = master_df[numeric_cols].corr().abs()
relevance = correlation_matrix.sum().sort_values()
print("Columnas con menor relevancia (correlación) con las demás:")
print(relevance)
```
